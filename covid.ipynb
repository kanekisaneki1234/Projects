```python
import pandas as pd
import numpy as np
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import pickle
```

```python
df = pd.read_csv('COVID.csv')
```

```python
df
```

```python
df["COVID"][df["COVID"]==1].count()
```

```python
indices = df.index[df['COVID'] == 1]
```

```python
indices
```

```python
np.random.seed(42)
random_indices = np.random.choice(indices, size=30, replace=False)
```

```python
random_indices
```

```python
for index in random_indices:
    # Replace the value in the 'City' column at the selected index with a new value
    df.at[index, 'COVID'] = 0
```

```python
df.loc[random_indices, "COVID"]
```

```python
df["COVID"][df["COVID"]==1].count()
```

```python
df["COVID"][df["COVID"]==0].count()
```

```python
df.shape
```

```python
train, test = train_test_split(df, test_size=0.3)
```

```python
print(train.shape)
```

```python
print(test.shape)
```

```python
train_x = train[['Gender', 'Nocomorbidity','Cardiovascular','Endocrine', 'tumor', 'Respiratory', 'Digestive', 'Renal', 'Liver', 'Fever', 'Cough', 'Chest tightness', 'Fatigue', 'Diarrhea', 'RNA']]
train_y = train.COVID
```

```python
test_x = train[['Gender', 'Nocomorbidity','Cardiovascular','Endocrine', 'tumor', 'Respiratory', 'Digestive', 'Renal', 'Liver', 'Fever', 'Cough', 'Chest tightness', 'Fatigue', 'Diarrhea', 'RNA']]
test_y = train.COVID
```

```python
train_x.head()
```

```python
train_y.head()
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(train_x, train_y)
```

```python
prediction = knn.predict(test_x)
```

```python
print("Accuracy:",metrics.accuracy_score(prediction, test_y))
```

```python
pickle.dump(knn,open('covid.pkl','wb'))
```
